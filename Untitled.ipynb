```python
import os
os.environ.setdefault('JAVA_HOME', '/opt/homebrew/opt/openjdk@11')   # optional if you use Homebrew Java
os.environ.setdefault('SPARK_LOCAL_IP', '127.0.0.1')                 # helps with macOS binding issues

from pyspark.sql import SparkSession

spark = (SparkSession.builder
         .appName("DataQuality")
         .config("spark.driver.bindAddress", "127.0.0.1")
         .config("spark.driver.host", "127.0.0.1")
         .getOrCreate())

print("Spark version:", spark.version)
df = spark.createDataFrame([(1, "a"), (2, "b")], ["id", "val"])
df.show()
spark.stop()

```
